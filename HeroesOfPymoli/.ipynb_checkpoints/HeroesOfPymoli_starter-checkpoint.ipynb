## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#print(purchase_data["Age"].max())
#print(purchase_data["Age"].min())

# Create the bins in which Data will be held
# Bins are 0, 60, 70, 80, 90, 100
bins = [0, 9, 14, 19, 24, 29, 34, 39, 46]
#bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]
#bins = [0, 11, 16, 21, 26, 31, 36, 41, 45]

# Create the names for the four bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age_Bin"] = pd.cut(purchase_data["Age"], bins, right=True, labels=group_labels, include_lowest=True)
purchase_data.head()

# Create a GroupBy object based upon "View Group"
#age_count = purchase_data['SN'].count()
age_group = purchase_data.groupby("Age_Bin")
age_count = age_group['SN'].count()

age_group_df = age_count.to_frame().reset_index()
#age_group_df

agesum = age_group_df["SN"].sum()
#agesum

age_group_df["Percentage of Players"] = (age_group_df["SN"] / agesum) *100
age_group_df["Percentage of Players"] = age_group_df["Percentage of Players"].map("{:.2f}".format)
#gender2_sort_df = gender2_df.sort_values("Percentage of Players", ascending = False)
age_group_df = age_group_df.rename(columns={"SN":"Total Counts"})
#gender2_sort_df.head()
age_group_df
```
